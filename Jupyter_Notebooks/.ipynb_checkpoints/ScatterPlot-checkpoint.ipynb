```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
sns.set_style("ticks")

df = pd.read_csv("G:\my drive\machine learning\Cholangiocarcinoma_ML\Data_Files\cc_cancer_mutations_post_LASSO.csv")
df.sort_values('cancer_type')
df['sample'] = df.index
df = df.melt(['cancer_type','sample'])
g = sns.relplot(
    data=df,
    x="variable", y="sample", hue="cancer_type", size="value",
    hue_norm=(-1, 1), palette='tab10',
    height=22, sizes=(1, 20), size_norm=(0, 1)
)
#g._legend.remove()
g.set(xlabel="Genes", ylabel="Samples");
```
